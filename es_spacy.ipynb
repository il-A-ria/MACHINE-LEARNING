```python
import spacy 
from spacy import displacy
```

```python
nlp = spacy.load('en_core_web_sm')
text = ("Evacuations Begin in Besieged City of Mariupol, Official Says. An evacuation effort from the port city was underway on Friday as residents awaited assistance that could provide relief to civilians trapped for weeks.A Russian official said that Ukraine fired on an oil depot in the city of Belgorod. Another round of peace talks began via video link. Here’s the latest.")
doc = nlp(text)

displacy.render(doc, style='ent', jupyter=True)
```

```python
for token in doc:
    print(token.text, token.pos_, token.dep_, token.lemma_)
```

```python
displacy.render(doc, style='dep', jupyter=True)
```

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')

text = ("Evacuations Begin in Besieged City of Mariupol, Official Says. An evacuation effort from the port city was underway on Friday as residents awaited assistance that could provide relief to civilians trapped for weeks.A Russian official said that Ukraine fired on an oil depot in the city of Belgorod. Another round of peace talks began via video link. Here’s the latest.")
doc = nlp(text)

print('Polarity: ', doc._.polarity)
print('Subjectivity: ', doc._.subjectivity)
print('Assessments: ', doc._.assessments)
```
