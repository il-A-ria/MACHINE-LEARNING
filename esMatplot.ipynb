```python
## ESERCITAZIONE LAB
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-4,4,100)  #quanti valori vedere sulla x?
y = np.linspace(-10,10,100)
plt.plot(x, y, '-r', label='y=2x+1')
plt.title('Graph of y=2x+1')
plt.xlabel('x', color='#1C2843' )
plt.ylabel('y', color='#1C2843')
plt.legend(loc='upper left')
#plt.grid()
plt.show()    #che significa?
```

```python
# 100 linearly spaced numbers
x = np.linspace(-6,5,100)
y = np.linspace(-50,50,100)    #perchè viene una parabola?
# the function, which is y = ??? here
y = x**2
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, '-g')
```
