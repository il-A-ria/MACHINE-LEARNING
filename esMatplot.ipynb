```python
## ESERCITAZIONE LAB
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5,5,100)  #quanti valori vedere sulla x?
y = 2*x+1

#plt.figure(figsize=)     #se vogliamo ingrandire. figsize deve essere la prima impostazione, se vogliamo metterla
plt.plot(x, y, '-r', label='y=2x+1')
plt.title('Graph of y=2x+1')
plt.xlabel('x', color='#1C2843' )
plt.ylabel('y', color='#1C2843')
plt.legend(loc='upper left')
#plt.grid()
plt.show()    
```

```python
# 100 linearly spaced numbers
x = np.linspace(-5,5,100)      #funzione dispari perchè non la si mette centrale
#perchè passa per l'origine? perchè l'intercetta è 0.
#l'equazione è x al cubo. perchè? i valori che ci sono sull'asse x al cubo, fanno
# più o meno i valori che ci sono nella funzione nell'asse y.

y = x**3

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, '-g')
```

```python
# 100 linearly spaced numbers

x = np.linspace(-np.pi,np.pi,100) #che funzione è? periodic, delimitata da -1 a 1
                                  #è un seno, perchè il s1n di 0 è 0.
                                  #il sin è max a 90 gradi, 1.
                                  #- p greco, pi greco (np.p)
                                  #se io ho un raggio o un diametro, quanto misura la circonferenza?
# the function, which is y = sin(x) here
y = np.cos(x)
                                  #la funzione periodica si ripete ogni periodo t. 
                                  #quanto vale questo periodo t? 2 volte pi greco/omega
                                  #omega è la pulsazione della funzione. 
                                  #preso un omega oìunitario si dirà che la funzione si ripete ogni 2 pi greco
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'b')
# show the plot
plt.show()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,100) 
# the function, which is y = sin(x) here
y = np.sin(x)
z = np.cos(x)
                                  
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'c', label='y=sin(x)')
plt.plot(x,z, 'm', label='y=cos(x)')
plt.legend(loc='upper left')
# show the plot
plt.show()
```

```python
x = np.linspace(-2,2,100) 

y = np.exp(x)
                                  
# setting the axes at the centre
fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'y', label='y=e^x')
plt.legend(loc='upper left')
# show the plot
plt.show()
```
